```python
# Statistical test for stationality: the augmented dicky-fuller's test
# Null hypothesis of dicky-fuller is that the time series is non-stationary due to trend
from statsmodels.tsa.stattools import adfuller
```

```python
web_traffic_df['visits'].plot()
plt.show()
```

```python
results = adfuller(web_traffic_df['visits'])
print("Test statistic: ", results[0]) # The more negative this number is
print("p-value: ", results[1])
print("Critical values: ", results[4])
```

```python
web_traffic_df.loc['2015']['visits'].plot()
plt.show()
```

```python
results = adfuller(web_traffic_df.loc['2015']['visits'])
print("Test statistic: ", results[0]) # The more negative this number is
print("p-value: ", results[1])
print("Critical values: ", results[4])
```

There seem to be a negative down trend in the 2015 data, which is why the p-value is so high. We could try to eliminate this by subtracting each data point by the value of the previous data point.

```python
web_traffic_df.loc['2015']['visits'].diff().dropna().plot()
plt.ylabel("$\Delta$ (daily visit)")
plt.show()
```

```python
results = adfuller(web_traffic_df.loc['2015']['visits'].diff().dropna())
print("Test statistic: ", results[0]) # The more negative this number is
print("p-value: ", results[1])
print("Critical values: ", results[4])
```

While this worked in this case, there could be times when you need to perform multiple differences or use an alternative approach.

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX
```

```python
from warnings import filterwarnings
filterwarnings('ignore')
```

```python
ar_order = []
ma_order = []
aic = []
bic = []
for p in range(5):
    for q in range(5):
        model = SARIMAX(web_traffic_df.loc['2015':'2016-03']['visits'], order=(p,1,q), seasonal_order=(1,1,1,7))
        results = model.fit()
        ar_order.append(p)
        ma_order.append(q)
        aic.append(results.aic)
        bic.append(results.bic)
```

```python
arima_results = pd.DataFrame({'ar_order':ar_order,
                              'ma_order':ma_order,
                              'aic':aic,
                              'bic':bic})

arima_results
```

```python
plt.figure(figsize=(20,8))
arima_results['aic'].plot()
arima_results['bic'].plot()
plt.xticks(range(25),list(zip(list(arima_results['ar_order']), list(arima_results['ma_order']))))
plt.xlabel("(ar, ma)")
plt.xticks(rotation=90)
plt.legend()
plt.show()
```

```python
training_df = web_traffic_df.loc['2015':'2016-03']['visits']
```

```python
training_df
```

```python
model = SARIMAX(training_df, order=(1, 1, 1), seasonal_order=(1,1,1,7))
```

```python
results = model.fit()
```

```python
# One step ahead forecast prediction (Testing forecast in sample)
forecast = results.get_prediction(start=-30)
```

```python
mean_forecast = forecast.predicted_mean
```

```python
mean_forecast
```

```python
confidence_interval = forecast.conf_int()
```

```python
confidence_interval.head()
```

```python
plt.figure(figsize=(15,10))

# Plot forecast
plt.plot(training_df, color='blue', label='time series')
plt.plot(mean_forecast, color='red', label='forecast')

# Plot uncertainty
plt.fill_between(confidence_interval.index, y1=confidence_interval.iloc[:,0], 
                 y2=confidence_interval.iloc[:,1], color='pink')

plt.legend()
plt.show()
```

```python
# Dynamic forecast prediction (Testing forecast in sample)
forecast = results.get_prediction(start=-14, dynamic=True)
```

```python
forecast.predicted_mean
```

```python
plt.figure(figsize=(15,10))

# Plot forecast
plt.plot(training_df, color='blue', label='time series')
plt.plot(forecast.predicted_mean, color='red', label='forecast')

# Plot uncertainty
plt.fill_between(forecast.conf_int().index, y1=forecast.conf_int().iloc[:,0], 
                 y2=forecast.conf_int().iloc[:,1], color='pink')

plt.legend()
plt.show()
```

```python
# Testing forecast ahead of sample
forecast = results.get_forecast(steps=14)
```

```python
plt.figure(figsize=(15,10))

# Plot forecast
plt.plot(training_df, color='blue', label='time series_training')
plt.plot(testing_df, color='green', label='time_series_testing')
plt.plot(forecast.predicted_mean, color='red', label='forecast')

# Plot uncertainty
plt.fill_between(forecast.conf_int().index, y1=forecast.conf_int().iloc[:,0], 
                 y2=forecast.conf_int().iloc[:,1], color='pink')

plt.legend()
plt.show()
```
